#### Display the trending Python repositories in GitHub

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/python?since=daily'
````

````python
#your code

html = requests.get(url).content

soup = BeautifulSoup(html, "lxml")
#soup

tags = ['h1']
text = [element.text for element in soup.find_all(tags)][1:]
#print(text)

new_text = []

for item in text:
    x = item.strip()
    new_text.append(x)

#print(new_text)

clean_text = []

for i in new_text:
    new = i.replace('\n','')
    new2 = new.replace('      ', ' ')
    clean_text.append(new2)
              
'''
L.S. Yes, very good!
'''        
        
print(clean_text)

````

#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page

````python
# This is the url you will scrape in this exercise
url ='https://en.wikipedia.org/wiki/Python' 

'''
L.S. It is good practice to import modules at the beginning of the notebook. 
'''

import re
````

````python
#your code

'''
L.S. Good!
'''

html = requests.get(url).content

soup = BeautifulSoup(html, "lxml")
#print(soup)

links  = []

for link in soup.find_all('a', attrs={'href': re.compile("http")}):
    links.append(link.get('href'))


print(links)
````

#### Top 10 languages by number of native speakers stored in a Pandas Dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
#your code

'''
L.S. Well done!
'''



html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")

table = soup.find_all('table')[0]
table

rows = table.find_all('tr')
rows

rows2 = [row.find_all('td') for row in rows]
rows2

text = [[x.text for x in row] for row in rows2]
#print(text)

text2 = []
for i in text:
    sub = []
    for x in i:
        y = x.rstrip('\n')
        sub.append(y)
    text2.append(sub)
    
#print(text2)

col_names = ['Rank', 'Language', 'Primary Country', 'Total Countries', 'Speakers', '% of the World population', 'Language Family']


lang_df = pd.DataFrame(text2, columns = col_names)
lang_df.head()

lang_df_2 = lang_df.drop(lang_df.index[0])
lang_df_2.head()
````
